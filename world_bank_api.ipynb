```python
# Dependencies
import requests
import pandas as pd
import json
import csv
import matplotlib.pyplot as plt
from scipy import stats
url = "http://api.worldbank.org/v2/"
print(url)
```

```python
# Get response and jsonify
response = requests.get(f'{url}lendingTypes?format=json').json()
print(json.dumps(response, indent = 4))
```

```python
#Extract lending categories
categories = [x['id'] for x in response[1]]
categories
```

```python
# Determine how many countries fall into each lending type.
country_count = {}
for i in categories:
    query = f"{url}countries?lendingType={i}&format=json"
    response = requests.get(query).json()
    print(json.dumps(response, indent = 4))
    country_count[i] = response[0]["total"]
country_count
```

```python
# Print the number of countries of each lending type
for key, value in country_count.items():
    print(f"There are {value} countries with lending type {key}")
```
